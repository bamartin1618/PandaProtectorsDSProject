```python
import pandas


"""
Our classification model is supposed to take in as input some features about a county (population density, number_of_deaths, area_sqml) and return as output the state that the county is in.


Output Labels: California and Nevada
"""


dataset = pandas.read_csv('us_county_sociohealth_data.csv')

dataset = dataset[(dataset['state'] == 'California') | (dataset['state'] == 'Nevada')]

input_features = dataset[['area_sqmi', 'num_deaths', 'population_density_per_sqmi']]
output_features = dataset['state']

for column_name in input_features.columns.values:
    
    input_features[column_name] = input_features[column_name].fillna(input_features[column_name].mean())
output_features

# 0 for california, 1 for nevada
# LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

output_labels = LabelEncoder().fit_transform(output_features)

train_x, test_x, train_y, test_y = train_test_split(input_features, output_labels, test_size = 0.2, random_state = 24)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

classifier = SVC()
classifier.fit(train_x, train_y)

predictions = classifier.predict(test_x)

print(f1_score(test_y, predictions))
```

```python
# Decision Trees
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier()
classifier.fit(train_x, train_y)

predictions = classifier.predict(test_x)

print(accuracy_score(test_y, predictions))
print(f1_score(test_y, predictions))
# K Nearest Neighbors
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier()
classifier.fit(train_x, train_y)

predictions = classifier.predict(test_x)

print(accuracy_score(test_y, predictions))
print(f1_score(test_y, predictions))
```
