```python
def iterative_list_sum(list_of_numbers): # Iterative solution for our problem
    
    total = 0
    
    for i in range(len(list_of_numbers)):
        total += list_of_numbers[i]
        
    return total

# recursive_list_sum(list_of_numbers)
# [1, 2, 3, 4, 5]
# 1 + sum([2, 3, 4, 5])
# 1 + 2 + sum([3, 4, 5])
# sum(list_of_numbers) = list_of_numbers[0] + sum(list_of_numbers[1:])

def recursive_list_sum(list_of_numbers):
    
    # if the list has zero numbers, then the sum is going to be zero
    # if the list has one number, then the sum will just be that number.
    
    if len(list_of_numbers) == 0:
        return 0
    if len(list_of_numbers) == 1:
        return list_of_numbers[0]
    
    return list_of_numbers[0] + recursive_list_sum(list_of_numbers[1:])
```

```python
"""
Euclidean Algorithm -

If a > b and we want to find the greatest common divisor of a and b -

gcd(a, b) = gcd(b, a % b)

lcm(a, b) = (a * b) / (gcd(a, b))
"""
import time

def gcd(a, b):
    
    if a < b:
        a, b = b, a
    if a % b == 0:
        return b
    
    return gcd(b, a % b)

def slow_gcd(a, b):
    
    divisors_of_a = []
    divisors_of_b = []
    
    for i in range(1, a+1):
        
        if a % i == 0:
            divisors_of_a.append(i)
            
    for i in range(1, b+1):
        
        if b % i == 0:
            divisors_of_b.append(i)
            
    maximum_divisor = 1
    
    for divisor in divisors_of_a:
        if divisor in divisors_of_b and divisor > maximum_divisor:
            maximum_divisor = divisor
    return maximum_divisor
```
